```python
# Scrape data from booking.com and store the data in the csv files.

url = "https://www.booking.cn/reviewlist.en-gb.html"

n=0
for page in range (0,589):
    params = {
    "aid": "956743",
    "sid": "98c68f2a43d7bba4e8f8115550840edf",
    "cc1": "fi",
    "dist": "1",
    "pagename": "radisson-sas-seaside",
    "srpvid": "b7694639a3dc02e5",
    "type": "total",
    "rows": "10",
    "_": "1678356515662",
    "offset": page*10
}

    response = requests.get(url, headers=headers, cookies=cookies, params=params)

    content = response.content.decode("utf8")
    soup = BeautifulSoup(content,"lxml")    

    results = soup.find_all("div",{"class":"c-review-block"})
    print (page+1, "page start, total 10 reviews")


    for result in results: 
        ## Review_No
        n=n+1
        Review_No = n

        ## Guest_Name
        try:
            Guest_Name = result.find("span",{"class":"bui-avatar-block__title"}).text.replace('\n','')
        except:
            Guest_Name = ""

        ## Guest_Region
        try:
            Guest_Region = result.find("span",{"class":"bui-avatar-block__subtitle"}).text.replace('\n','')
        except:
            Guest_Region = ""
            
        ## Reviewed_Date
        try:
            Reviewed_Date = result.find_all("span",{"class":"c-review-block__date"})[1].text.replace("\n","")
        except:
            Reviewed_Date = ""
        
        ## Review_Rating
        try:
            Review_Rating = result.find("div",{"class":"bui-review-score__badge"}).text
        except:
            Review_Rating = ""

         ## Room_Type
        if Guest_Name != "Anonymous":
            try:
                Room_Type = result.find_all("div",{"class":"bui-list__body"})[0].text.replace('\n','')
            except:
                Room_Type = ""

            ## "Night_Stayed"
            try:
                Night_Stayed = result.find_all("div",{"class":"bui-list__body"})[1].text.split("")[0].replace('\n','').replace("night","").replace("s","").replace(" ","")
            except:
                Night_Stayed = ""

            ## "Month_of_Stay"
            try:
                Month_of_Stay = result.find_all("div",{"class":"bui-list__body"})[1].text.split("")[1].replace('\n','')
            except:
                Month_of_Stay = ""

            ## "Stay_Type"
            try:
                Stay_Type = result.find_all("div",{"class":"bui-list__body"})[2].text.replace('\n','')
            except:
                Stay_Type = ""
                
        else:
            try:
                Room_Type = ""
            except:
                Room_Type = ""

            ## "Night_Stayed"
            try:
                Night_Stayed = result.find_all("div",{"class":"bui-list__body"})[0].text.split("")[0].replace('\n','').replace("night","").replace("s","").replace(" ","")
            except:
                Night_Stayed = ""

            ## "Month_of_Stay"
            try:
                Month_of_Stay = result.find_all("div",{"class":"bui-list__body"})[0].text.split("")[1].replace('\n','')
            except:
                Month_of_Stay = ""

            ## "Stay_Type"
            try:
                Stay_Type = result.find_all("div",{"class":"bui-list__body"})[1].text.replace('\n','')
            except:
                Stay_Type = ""


        #"Review_Details"
        try:
            Review_title = result.find("h3",{"class":"c-review-block__title c-review__title--ltr"}).text.replace('\n',' ')
        except:
            Review_title =  ""
        try:
            Review_like = result.find("div",{"class":"c-review__row"}).text.replace('\n',' ')
        except:
            Review_like = ""
        try:
            Review_dislike = result.find("div",{"class":"c-review__row lalala"}).text.replace('\n',' ')
        except:
            Review_dislike = ""

        Review_Details = Review_title +Review_like+Review_dislike
        # print (Review_Details)
        
        csv_writer.writerow([Review_No,page+1,Review_Rating, Guest_Name,Guest_Region,Reviewed_Date,Room_Type,Night_Stayed,Month_of_Stay,Stay_Type,Review_Details])
        print("number: ",n, "review in", "page: ",page+1)
        
        
    print (page+1,"done", n,"reviews", "done, ", 5891-n,"left." ) 
    


csv_file.close()
print ("done")


```
